```python
# Dependencies
import pandas as pd
import sqlite3
import csv
from pandas import DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Store filepaths in variable
fortune = "fortune1000-final.csv"
employees = "Employees_By_State.csv"
number = "Num_In_State.csv"
profit = "Profit_By_State.csv"
revenue = "Revenue_By_State.csv"
```

```python
# Read our Data files with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
fortune_df = pd.read_csv(fortune, encoding="ISO-8859-1")
employees_df = pd.read_csv(employees, encoding="ISO-8859-1")
number_df = pd.read_csv(number, encoding="ISO-8859-1")
revenue_df = pd.read_csv(revenue, encoding="ISO-8859-1")
profit_df = pd.read_csv(profit, encoding="ISO-8859-1")
```

```python
###  only necessary columns
fortune_df = fortune_df[["rank", "title", "Revenues ($M)", "Profits ($M)", "Employees", "Sector", "Industry", "Latitude", "Longitude",]]
fortune_df.head()
```

```python
fortune500_df = fortune_df.drop(fortune_df.index[500:1000])
```

```python
fortune500_df
```

```python
# Create Fortune 500 Database
conn = sqlite3.connect("fortune500.db")
```

```python
##push the dataframe to sql 
fortune500_df.to_sql("fortune500_df", conn, if_exists="replace")

##create the fortune 500 table

conn.execute(
    """
    create table if not exists fortune500_table as 
    select * from fortune500_df 
    """)

##push the dataframe to sql 
employees_df.to_sql("employees_df", conn, if_exists="replace")

##create the employee table

conn.execute(
    """
    create table if not exists employee_table as 
    select * from employees_df 
    """)

##push the dataframe to sql 
number_df.to_sql("number_df", conn, if_exists="replace")

##create the numbers table

conn.execute(
    """
    create table if not exists number_table as 
    select * from number_df 
    """)

##push the dataframe to sql 
revenue_df.to_sql("revenue_df", conn, if_exists="replace")

##create the revenue table

conn.execute(
    """
    create table if not exists revenue_table as 
    select * from revenue_df 
    """)

##push the dataframe to sql 
profit_df.to_sql("profit_df", conn, if_exists="replace")

##create the profit table

conn.execute(
    """
    create table if not exists profit_table as 
    select * from profit_df 
    """)
```

```python
# Display data in the table

c = sqlite3.connect('fortune500.db')
cur = c.cursor()
cur.execute("SELECT * from fortune500_table")
fortune500_table = cur.fetchall()
```
