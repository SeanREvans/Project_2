```python
# Dependencies
import pandas as pd
import sqlite3
import csv
from pandas import DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
fortune = "fortune1000-final.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
fortune_df = pd.read_csv(fortune, encoding="ISO-8859-1")
```

```python
# Show just the header
fortune_df.head()
```

```python
###  only necessary columns
fortune_df = fortune_df[["rank", "title", "Revenues ($M)", "Profits ($M)", "Employees", "Sector", "Industry", "Latitude", "Longitude",]]
fortune_df.head()
```

```python
# Create Database
conn = sqlite3.connect("fortune.db")
```

```python
##push the dataframe to sql 
fortune_df.to_sql("fortune_df", conn, if_exists="replace")

##create the table

conn.execute(
    """
    create table if not exists fortune1000_table as 
    select * from fortune_df 
    """)
```

```python
# Display data in the table

c = sqlite3.connect('fortune.db')
cur = c.cursor()
cur.execute("SELECT * from fortune1000_table limit 500")
fortune500_view = cur.fetchall()
```

```python
for row in fortune500_view:
        print(row)
```
